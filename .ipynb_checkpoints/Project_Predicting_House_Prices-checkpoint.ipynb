# LINEAR REGRESSION

```python
data.isnull().sum()
#Make sure no null values before building a model
```

```python
data.dropna(inplace = True)
data.isnull().sum()
```

```python
X = data.drop(columns=['Sale Price'])
y = data['Sale Price']
```

```python
#VARIABLE TRANSFORMATION
#Plot Histogram for all the Numerical Values
numerical_columns = ['No of Bedrooms', 'No of Bathrooms', 'Lot Area (in Sqft)',
       'No of Floors',
       'Area of the House from Basement (in Sqft)', 'Basement Area (in Sqft)',
       'Age of House (in Years)', 'Latitude', 'Longitude',
       'Living Area after Renovation (in Sqft)',
       'Lot Area after Renovation (in Sqft)',
       'Years Since Renovation']
#These also contains float values - Convert them to int

for i in numerical_columns:
    X[i] = pd.to_numeric(X[i])
```

```python
def hist(data, variables):
    plt.figure(figsize = (len(variables)*6,6), dpi = 120)
    for i,j in enumerate(variables):
        plt.subplot(1,len(variables),i+1)
        plt.hist(data[j])
        plt.title(j)
hist(X , numerical_columns)
```

```python
#Obverse the right skew from the diagrams and apply log
#If left skew them square the values


right_skew_variables = ['No of Bedrooms', 'No of Bathrooms', 'Lot Area (in Sqft)',
       'No of Floors',
       'Area of the House from Basement (in Sqft)', 'Basement Area (in Sqft)',
        'Longitude',
       'Living Area after Renovation (in Sqft)',
       'Lot Area after Renovation (in Sqft)',
       'Years Since Renovation']

def right_skew(value):
  return np.log(abs(value+500))
```

```python
#To apply log to all elements = Remove right skew
for i in right_skew_variables:
  X[i] = X[i].map(right_skew)
```

```python
#Check again for skew
hist(X , numerical_columns)
```

```python
#Remove Infinte Values  - log(0) is infinite
#remove infinite with np.nan and drop them
X = X.replace(np.inf, np.nan)
X.dropna(inplace=True)
```

```python
#Now you wil get exact graph
hist(X , numerical_columns)
```

# COLLINEARITY

```python
X.corr()
```

```python
#If collinearity > 0.5 remove them
## pair of independent variables with correlation greater than 0.5
k = X.corr()
z = [[str(i),str(j)] for i in k.columns for j in k.columns if (k.loc[i,j] >abs(0.5))&(i!=j)]
z, len(z)
```

```python
X.isnull().sum()
```

```python
np.isinf(X).any(1)
```

```python
#Removing can be done by VIF(Variance Inflation Factor) if VIF > 5

# Importing Variance_inflation_Factor funtion from the Statsmodels
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif_data = X[:]

## Calculating VIF for every column
VIF = pd.Series([variance_inflation_factor(vif_data.values, i) for i in range(vif_data.shape[1])], index = vif_data.columns)
VIF
```

```python
#Remove those columns whose VIF is > 5
def MC_remover(data):
  vif = pd.Series([variance_inflation_factor(data.values, i) for i in range(data.shape[1])], index = data.columns)
  if vif.max() > 5:
    print(vif[vif == vif.max()].index[0],'has been removed')
    data = data.drop(columns = [vif[vif == vif.max()].index[0]])
    return data
  else:
    print('No Multicollinearity present anymore')
    return data
```

```python
for i in range(7):
  vif_data = MC_remover(vif_data)

vif_data.head()
```

```python
VIF = pd.Series([variance_inflation_factor(vif_data.values, i) for i in range(vif_data.shape[1])], index = vif_data.columns)
VIF, len(vif_data.columns)
```

```python
X = vif_data[:]
```

# TRAIN AND TEST THE MODEL

```python
#Data Cleaning is completed ,split the data and train the model
from sklearn.model_selection import train_test_split
X_train , X_test , y_train , y_test = train_test_split(X,y,test_size=0.3,random_state=101)
X_train.shape , X_test.shape , y_train.shape , y_test.shape
```

```python
#LINEAR REGRESSION
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X_train, y_train)
```

```python
predictions = lin_reg.predict(X_test)
predictions
```

```python
#Y = x1*w1 + x2*w2 + x3*w3 + x4*w4 + .............
#w1,w2,w3...... are coefficients of eqn
lin_reg.coef_
```

```python
lin_reg.score(X_test, y_test)
```

# RESIDUALS

```python
#Compare true values with predicted values
residuals = predictions - y_test
residual_table = pd.DataFrame({'residuals':residuals,'predictions':predictions})
residual_table
```

```python
#Low to High 
residual_table = residual_table.sort_values( by = 'predictions')
```

```python
z = [i for i in range(int(residual_table['predictions'].max()))]
k = [0 for i in range(int(residual_table['predictions'].max()))]
```

```python
plt.figure(dpi = 130, figsize = (17,7))

plt.scatter( residual_table['predictions'], residual_table['residuals'], color = 'red', s = 2)
plt.plot(z, k, color = 'green', linewidth = 3, label = 'regression line')
plt.ylim(-800000, 800000)
plt.xlabel('fitted points (ordered by predictions)')
plt.ylabel('residuals')
plt.title('residual plot')
plt.legend()
plt.show()
```

```python
plt.figure(dpi = 100, figsize = (10,7))
plt.hist(residual_table['residuals'], color = 'red', bins = 200)
plt.xlabel('residuals')
plt.ylabel('frequency')
plt.title('distribution of residuals')
plt.show()
```

```python
coefficients_table = pd.DataFrame({'column': X_train.columns,'coefficients': lin_reg.coef_})
coefficient_table = coefficients_table.sort_values(by = 'coefficients')
```

```python
plt.figure(figsize=(8, 6), dpi=120)
x = coefficient_table['column']
y = coefficient_table['coefficients']
plt.barh( x, y)
plt.xlabel( "Coefficients")
plt.ylabel('Variables')
plt.title('Normalized Coefficient plot')
plt.show()
```
